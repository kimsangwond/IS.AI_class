```python
import pandas as pd
```

```python
data = pd.read_csv("lottery.csv", sep=",")
```

```python
data
```

```python
df = pd.DataFrame( data, columns = ["first", "second", "third", "fourth", "fifth", "sixth", "bonus"])
```

```python
first_df=df['first'].value_counts()
```

```python
second_df=df['second'].value_counts()
```

```python
third_df=df['third'].value_counts()
```

```python
fourth_df=df['fourth'].value_counts()
```

```python
fifth_df=df['fifth'].value_counts()
```

```python
sixth_df=df['sixth'].value_counts()
```

```python
bonus_df=df['bonus'].value_counts()
```

```python
df_matrix = pd.concat([first_df, second_df, third_df, fourth_df, fifth_df, sixth_df, bonus_df], join='outer', axis=1)
```

```python
df_matrix=df_matrix.fillna(0)
```

```python
df_matrix['sum']=df_matrix['first']+df_matrix['second']+df_matrix['third']+df_matrix['fourth']+df_matrix['fifth']+df_matrix['sixth']+df_matrix['bonus']
```
